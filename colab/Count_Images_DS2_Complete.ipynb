```python
# DS2 Complete: PNG sayıları (suffix tabanlı: _seg/_inst)
import os

ROOT = '/content/drive/MyDrive/omr_dataset/dataset/ds2/ds2_complete'
print({'ROOT_exists': os.path.exists(ROOT)})

# Kök altındaki tüm .png dosyalarını say; suffix ile sınıflandır

def count_png_suffix(root):
    total_png = 0
    seg = 0
    inst = 0
    for r, _, files in os.walk(root):
        for f in files:
            if os.path.splitext(f)[1].lower() == '.png':
                total_png += 1
                fl = f.lower()
                if fl.endswith('_seg.png'):
                    seg += 1
                elif fl.endswith('_inst.png'):
                    inst += 1
    images = total_png - seg - inst
    return images, seg, inst

images_total, segmentation_total, instance_total = count_png_suffix(ROOT)
print({'images_total': images_total, 'segmentation_total': segmentation_total, 'instance_total': instance_total})
```
